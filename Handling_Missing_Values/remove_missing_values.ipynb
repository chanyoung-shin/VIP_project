```python
import pandas as pd
import numpy as np

# 1. 추출된 데이터 불러오기
```

```python
file_path = "extracted_variables_with_merged_cancers.csv"
df = pd.read_csv(file_path)

```

```python

vars_789 = [
    'SEX_A', 'HISDETP_A', 'RACEALLP_A', 'PHSTAT_A', 'CANEV_A', 
    'SUPPORT_A', 'LONELY_A', 'SLPREST_A', 'ANXFREQ_A', 'DEPFREQ_A', 
    'FDSBALANCE_A', 'MODTPR_A', 'VIGTPR_A', 
    'BREASCAN_A', 'PROSTCAN_A', 'SKNMCAN_A', 'SKNNMCAN_A']

vars_789 = [var for var in vars_789 if var in df.columns]
for var in vars_789:
    df[var] = df[var].replace([7, 8, 9], np.nan)


vars_97_99 = [
    'EDUCP_A', 'SLPHOURS_A', 'DRKAVG12M_A', 'DRK12MWK_A', 
    'BREASAGETC_A', 'PROSTAGETC_A', 'SKNMAGETC_A', 'SKNNMAGETC_A', 
    'Merged_Other_Cancer_Age']
vars_97_99 = [var for var in vars_97_99 if var in df.columns]
for var in vars_97_99:
    df[var] = df[var].replace([97, 98, 99], np.nan)


if 'AGEP_A' in df.columns:
    df['AGEP_A'] = df['AGEP_A'].replace([997, 998, 999], np.nan)

vars_9997_9999 = ['MODNR_A', 'VIGNR_A']
vars_9997_9999 = [var for var in vars_9997_9999 if var in df.columns]
for var in vars_9997_9999:
    df[var] = df[var].replace([9995, 9996, 9997, 9998, 9999], np.nan)


if 'SMKCIGST_A' in df.columns:
    df['SMKCIGST_A'] = df['SMKCIGST_A'].replace(9, np.nan) 

if 'FDSCAT3_A' in df.columns:
    df['FDSCAT3_A'] = df['FDSCAT3_A'].replace(8, np.nan) 




```

```python

if 'DRK12MWK_A' in df.columns:
    df['DRK12MWK_A'] = df['DRK12MWK_A'].replace(95, 0)


exercise_vars = ['MODFREQW_A', 'VIGFREQW_A']
exercise_vars = [var for var in exercise_vars if var in df.columns]
for var in exercise_vars:

    df[var] = df[var].replace(94, 0)

    df[var] = df[var].replace([95, 96, 97, 98, 99], np.nan)




output_file = "cleaned_nhis_data_nan.csv"
df.to_csv(output_file, index=False)
```

```python
import pandas as pd
import numpy as np


from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer


file_path = "cleaned_nhis_data_nan.csv"
df = pd.read_csv(file_path)


impute_vars = [
    'AGEP_A', 'SLPHOURS_A', 'DRKAVG12M_A', 
    'MODFREQW_A', 'VIGFREQW_A', 'MODNR_A', 'VIGNR_A',
    'PHSTAT_A', 'FDSCAT3_A'
]


impute_vars = [var for var in impute_vars if var in df.columns]
df_to_impute = df[impute_vars].copy()

print("MICE 보간 전 결측치 개수:")
print(df_to_impute.isnull().sum())


mice_imputer = IterativeImputer(
    max_iter=10,        
    random_state=42,    
    min_value=0      
)


print("\nImputing missing values ​​for lifestyle habits and basic information.")
imputed_data = mice_imputer.fit_transform(df_to_impute)


df_imputed = pd.DataFrame(imputed_data, columns=impute_vars).round(0)


df[impute_vars] = df_imputed

print("\nNumber of missing values ​​after MICE imputation (Cancer Diagnosis Age left untouched):")
print(df[impute_vars].isnull().sum())


output_file = "final_nhis_data_mice_imputed_corrected.csv"
df.to_csv(output_file, index=False)
```
